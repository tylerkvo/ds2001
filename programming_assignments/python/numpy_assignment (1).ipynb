**SIMULATION EXERCISES**

13) **(1 PT)** Draw 100 random values from a standard normal distribution, using SEED = 314.

```python
from numpy.random import seed
from numpy.random import normal

seed(314)

dev = normal(loc=0, scale=1, size=100)

print(dev)

# compute cumulative sum
# https://mathworld.wolfram.com/CumulativeSum.html
cumsum = np.cumsum(dev)
print(cumsum)
# plot the cumulative sum (called a random walk)
plt.plot(cumsum)
```

14) **(1 PT)** Write code to generate and plot 10 random walks. For reproducibility, explicitly state the seeds.  
Hints: 1) You can loop over the seeds 2) Calling *plt.plot()* in a loop will overplot on one graph

```python
from numpy.random import seed
from numpy.random import normal



count = 0
seed_count = 1
while count < 10:
    seed(seed_count)
    dev = normal(loc=0, scale=1, size=100)
    plt.plot(np.cumsum(dev))
    count += 1
    seed_count += 1
```
